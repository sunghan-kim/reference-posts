# 2. Quick Start

````python
import FinanceDataReader as fdr
````

<br>

## 2.1 Apple

````python
# Apple(APPL), 2017-01-01 ~ Now
df = fdr.DataReader('AAPL', '2017')
df.head()
````

````python
df.tail()
````

## 2.2 Amazon

````python
# AMAZON(AMZN), 2017
df = fdr.DataReader('AMZN', '2017-01-01', '2017-12-31')
df.head()
````

````python
df.tail()
````

## 2.3 Samsung

````python
# Samsung(005930), 1992-01-01 ~ 2018-12-31
df = fdr.DataReader('005930', '1992-01-01', '2018-12-31')
df.head()
````

````python
df.tail()
````

## 2.4 Country code

````python
# country code
#  ex) 000150: Doosan(KR), Yihua Healthcare(CN)
````

````python
# default: 'KR'
df = fdr.DataReader('000150', '2018-01-01', '2018-12-31')
df.head()
````

````python
df.tail()
````

````python
# country='KR'
df = fdr.DataReader('000150', '2018-01-01', '2018-12-31', country='KR')
df.head()
````

````python
df.tail()
````

````python
# country='CN'
df = fdr.DataReader('000150', '2018-01-01', '2018-12-31', country='CN')
df.head()
````

````python
df.tail()
````
