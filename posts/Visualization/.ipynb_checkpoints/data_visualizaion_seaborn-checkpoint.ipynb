**`Seaborn`**

- `Matplotlib`을 기반으로 다양한 색상 테마와 통계용 차트 등의 기능을 추가한 시각화 패키지

# 0. Summary

**1차원 실수 데이터**

$\Rightarrow$ `distplot(x)`
- 히스토그램

**1차원 카테고리 데이터**

$\Rightarrow$ `countplot(x, data)`
- 막대 그래프

**2차원 실수 데이터**

$\Rightarrow$ `jointplot(x, y, data, kind)`
- `kind='scatter'` : 산점도
- `kind='kde'` : 커널 밀도 히스토그램

**다차원(3차원이상) 실수 데이터**

$\Rightarrow$ `pairplot(df, hue)`

**2차원 카테고리 데이터**

$\Rightarrow$ `heatmap(data, cmap, annot, fmt)`

**2차원 복합 데이터**

`barplot(x, y, data)`
- `x` : 카테고리 데이터 (or 실수 데이터)
- `y` : 실수 데이터 (or 카테고리 데이터)
- `data` : 데이터프레임  
  
  
- 카테고리값에 따른 실수 값의 평균과 편차를 표시    
  
  
`boxplot(x, y, data, whis)`
- `x` : 카테고리 데이터 (or 실수 데이터)
- `y` : 실수 데이터 (or 카테고리 데이터)
- `data` : 데이터프레임  
- `whis` : 미지정 시 outlier 표시
  - `whis=np.inf` : outlier 표시 안함  
  
  
- 최소값, 1사분위수, 중앙값, 3사분위수, 최대값 이상치 표시  
  
  
`violinplot(x, y, data ,inner)`
- `x` : 카테고리 데이터 (or 실수 데이터)
- `y` : 실수 데이터 (or 카테고리 데이터)
- `data` : 데이터프레임  
- `inner` : 미지정 시 가운데 선 표출
  - `inner=None` : 가운데 선 표시 안함  
  
  
- 실수 데이터에 대해 세로 방향으로 커널 밀도 히스토그램을 그려줌  
  
  
`stripplot(x=, y=, data, jitter)`
- `x` : 카테고리 데이터 (or 실수 데이터)
- `y` : 실수 데이터 (or 카테고리 데이터)
- `data` : 데이터프레임
- `jitter` : 가로축상의 위치 조절
  - `jitter=True` : 가로축상의 위치를 무작위로 조정
  - `jitter=Flase` : 가로축상의 위치를 겹쳐서 하나의 직선위에 표출  
  
  
- 카테고리 데이터별 실수 데이터의 분포를 점으로 표시  
  
  
`swarmplot(x, y, data)`
- `x` : 카테고리 데이터 (or 실수 데이터)
- `y` : 실수 데이터 (or 카테고리 데이터)
- `data` : 데이터프레임  
  
  
- `stripplot`과 유사
- 데이터를 나타내는 점이 겹치지 않도록하여 표출

**다차원 복합 데이터**

- 2차원 복합 데이터를 표현하는 데 사용된 함수들을 그대로 사용  
(`barplot`,`boxplot`,`violinplot`,`stripplot`,`swarmplot`)
- `hue` 옵션과 `split`, `dodge` 옵션을 사용할 수 있음  
  
  
`hue`
- 카테고리 데이터 지정
- `y`에 지정한 카테고리 변수와 `hue`에 지정한 카테고리 변수 각각의 값에 의한 실수값의 변화를 보여줌
- 함수의 종류에 따라 `hue`값이 시각화되는 방법은 다름  
  
  
`split`, `dodge`
- `violinplot`, `stripplot`, `swarmplot` 함수에서 사용할 수 있는 옵션
- `hue`를 지정하므로써 2 가지 카테고리 변수 값에 따른 실수값의 분포를 보여줄 때 표현 방식에 변화를 주는 옵션  
  
  
`heatmap(data, annot, fmt, linewidth)`
- 두개의 카테고리 값에 의한 실수 값 변화 확인

**catplot**

`catplot(x, y, hue, row, data, kind, split, height, aspect)`
- `x` : 카테고리 데이터 (or 실수 데이터)
- `y` : 실수 데이터 (or 카테고리 데이터)
- `hue` : 카테고리 데이터, 색상으로 구분
- `row` : 카테고리 데이터, 행으로 구분,
- `data` : 데이터프레임
- `kind` : 그래프의 종류 (`violin`, `strip`, `swarm`)
- `split` : hue에 지정한 카테고리 값의 표현 방식 분리(`swarm`에서는 `dodge`)
- `height` : 각각의 row별 그래프의 높이
- `aspect` : 각각의 row별 그래프의 너비

# 1. 스타일

- `Seaborn`을 임포트하면 색상 등을 `Matplotlib`에서 제공하는 기본 스타일이 아닌 `Seaborn`에서 지정한 기본 스타일로 바꿈
- 따라서 동일한 `Matplotlib` 명령을 수행해도 `Seaborn`을 임포트 한 것과 하지 않은 플롯의 모양은 다르다.
- 자세한 내용은 다음 [문서](http://seaborn.pydata.org/tutorial/aesthetics.html)를 참조

## 1.1 `set`

- `set` 명령으로 색상, 틱 스타일 등 전반적인 플롯 스타일을 `Seaborn` 기본 스타일로 바꾼다.
- [set 관련 문서](http://seaborn.pydata.org/generated/seaborn.set.html) 참고

## 1.2 `set_style`

- 틱 스타일만 바꿀 수 있다.
- `darkgrid`, `whitegrid`, `dark`, `white`, `ticks` 스타일을 제공
- [set_style 관련 문서](http://seaborn.pydata.org/generated/seaborn.set_style.html) 참고
- [set_color_codes 관련 문서](http://seaborn.pydata.org/generated/seaborn.set_color_codes.html) 참고

```python
import seaborn as sns
sns.set()
sns.set_style('whitegrid')
sns.set_color_codes()

import matplotlib as mpl
mpl.rc('font', family='nanumgothic')
mpl.rc('axes', unicode_minus=False)

import matplotlib.pyplot as plt
%matplotlib inline

import numpy as np
```

# 3. 1차원 분포 플롯

- 1차원 데이터가 **실수**값 $\Rightarrow$ 히스토그램과 같은 **실수 분포 플롯**으로 나타냄
- 1차원 데이터가 **카테고리**값 $\Rightarrow$ **카운트 플롯**으로 나타냄

```python
iris = sns.load_dataset("iris") # 붓꽃 데이터
titanic = sns.load_dataset("titanic") # 타이타닉호 데이터
tips = sns.load_dataset("tips") # 팁 데이터
flights = sns.load_dataset("flights") # 여객운송 데이터
```

## 3.1 1차원 실수 분포 플롯

- 실수 분포 플롯은 **자료의 분포**를 묘사하기 위한 것
- `Matplotlib`의 단순한 히스토드램과 달리 **커널 밀도(Kernel Density)** 및 **러그(rug)** 표시 기능 및 다차원 복합 분포 기능 등을 제공
- 1차원 실수 분포 플롯 명령
  - `rugplot`
  - `kdeplot`
  - `distplot`

### 3.1.1 러그(rug) 플롯

- 러그(rug) 플롯은 데이터 위치를 x축 위에 작은 선분(rug)으로 나타내어 실제 데이터들의 위치를 보여줌
- [rugplot 관련 문서](http://seaborn.pydata.org/generated/seaborn.rugplot.html)

```python
x = iris.petal_length.values
```

```python
sns.rugplot(x)
plt.title("Iris 데이터 중, 꽃잎의 길이에 대한 Rug Plot")
plt.show()
```

### 3.1.2 커널 밀도(kernel density) 플롯

- **커널 밀도(kernel density)** : 커널이라는 함수를 겹치는 방법
- 히스토그램보다 부드러운 형태의 분포 곡선을 보여주는 방법
- [kdeplot 관련 문서](http://seaborn.pydata.org/generated/seaborn.kdeplot.html)
- 커널 밀도 추정에 대한 자세한 내용은 [scikit-learn 패키지](http://scikit-learn.org/stable/modules/density.html)를 참조

```python
sns.kdeplot(x)
plt.title("Iris 데이터 중, 꽃잎의 길이에 대한 Kernel Density Plot")
plt.show()
```

### 3.1.3 distplot 플롯

- `Seaborn`의 `distplot` 명령은 러그와 커널 밀도 표시 기능이 있음
- 그렇기 때문에 `Matplotlib`의 `hist` 명령보다 많이 사용됨
- [distplot 관련 문서](http://seaborn.pydata.org/generated/seaborn.distplot.html)

```python
sns.distplot(x, kde=True, rug=True)
plt.title("Iris 데이터 중, 꽃잎의 길이에 대한 Dist Plot")
plt.show()
```

## 3.2 카운트 플롯 (`countplot`)

- `countplot` 명령을 사용하면 각 카테고리 값별로 데이터가 얼마나 있는 지 표시할 수 있음
- [countplot 관련 문서](http://seaborn.pydata.org/generated/seaborn.countplot.html)
- `countplot` 명령은 **데이터프레임에만 사용**할 수 있음

**사용방법**

`countplot(x="column_name", data=dataframe)`
  - `data` : 대상이 되는 데이터프레임 지정
  - `x` : 데이터프레임의 열 이름 문자열 입력

```python
sns.countplot(x="class", data=titanic)
plt.title("타이타닉호의 각 클래스별 승객 수")
plt.show()
```

```python
sns.countplot(x="day", data=tips)
plt.title("요일별 팁을 준 횟수")
plt.show()
```

# 4. 다차원 데이터

- 데이터 변수가 여러 개인 다차원 데이터는 데이터의 종류에 따라 다음과 같은 경우가 있을 수 있다.
  - 분석하고자 하는 데이터가 모두 **실수값**인 경우
  - 분석하고자 하는 데이터가 모두 **카테고리값**인 경우
  - 분석하고자 하는 데이터에 **실수값**과 **카테고리값**이 섞여 있는 경우

## 4.1 2차원 실수형 데이터

- 데이터가 2차원이고 모두 연속적인 실수값  
$\Rightarrow$ **산점도(Scatter plot)**을 사용
- 산점도를 그리기 위해서는 `Seaborn` 패키지의 `jointplot` 명령을 사용
- `jointplot` 명령은 산점도 뿐만 아니라 차트의 가장자리(margin)에 각 변수의 히스토그램도 보여준다.
- [jointplot 관련 문서](http://seaborn.pydata.org/generated/seaborn.jointplot.html)
- `jointplot` 명령도 **데이터프레임만 사용**할 수 있다.

**사용방법**

`jointplot(x="x_name", y="y_name", data=dataframe, kind="scatter")`
- `data` : 대상이 되는 데이터프레임 지정
- `x` : x 변수가 될 데이터프레임의 열 이름 문자열
- `y` : y 변수가 될 데이터프레임의 열 이름 문자열
- `kind` : 차트의 종류 지정
  - `kind='scatter'` : 산점도

```python
sns.jointplot(x="sepal_length", y="sepal_width", data=iris)
plt.title("꽃받침의 길이와 넓이의 Joint Plot")
plt.show()
```

- `kind='kde'` 로 지정하면 커널 밀도 히스토그램을 그린다.

```python
sns.jointplot(x='sepal_length', y='sepal_width',
              data=iris,
              kind='kde')
plt.title("꽃받침의 길이와 넓이의 Joint Plot 과 Kernel Density Plot")
plt.show()
```

## 4.2 다차원 실수형 데이터

- 만약 3차원 이상의 데이터라면 `Seaborn` 패키지의 `pairplot` 명령을 사용
- `pairplot`은 데이터프레임을 인수로 받아 그리드(grid) 형태로 각 데이터 열의 조합에 대해 산점도를 그린다.
- 같은 데이터가 만나는 대각선 영역에는 해당 데이터의 히스토그램을 그린다.
- 이 그림을 산점도행렬이라고도 한다.
- [pairplot 관련 문서](http://seaborn.pydata.org/generated/seaborn.pairplot.html)

```python
sns.pairplot(iris)
plt.title("Iris Data의 Pair Plot")
plt.show()
```

- 만약 카테고리형 데이터가 섞여 있는 경우에는 `hue` 인수에 카테고리 변수 이름을 지정하여 카테고리 값에 따라 색상을 다르게 할 수 있다.
- `markers` 옵션은 각 포인트의 모양을 다르게 지정할 수 있다.

```python
sns.pairplot(iris, hue="species", markers=["o","s","D"])
plt.title("Iris Pair Plot, Hue로 꽃의 종을 시각화")
plt.show()
```

## 4.3 2차원 카테고리 데이터

- 만약 데이터가 2차원이고 모든 값이 카테고리 값이면 `heatmap` 명령을 사용한다.
- [heatmap 관련 문서](http://seaborn.pydata.org/generated/seaborn.heatmap.html)

**사용방법**

`heatmap(data, cmap, annot, fmt)`
- `data` : 2차원 카테고리 데이터
- `cmap` : 범주별 색상 구분
- `annot` : 값 표시 여부(annotation:주석)
  - `annot=True` : 값 표시
  - `annot=False` : 값 표시 안함
- `fmt` : 값의 자료형
  - `fmt='d'` : 정수
  - `fmt='f'` : 실수

```python
titanic_size = titanic.pivot_table(index="class", 
                                   columns="sex", 
                                   aggfunc="size")
titanic_size
```

```python
sns.heatmap(titanic_size,
            cmap=sns.light_palette("gray", as_cmap=True),
            annot=True,
            fmt="d")
```

## 4.4 2차원 복합 데이터

- 만약 데이터가 2차원이고 실수값, 카테고리값이 섞여 있다면 기존의 플롯 이외에도 다음과 같은 분포 플롯들을 이용할 수 있다.
  - `barplot`
  - `boxplot`
  - `pointplot`
  - `violinplot`
  - `stripplot`
  - `swarmplot`

### 4.4.1 `barplot`

- `barplot`은 카테고리 값에 따른 실수 값의 평균과 편차를 표시하는 기본적인 바 차트를 생성
- 막대의 높이 $\Rightarrow$ 평균
- 에러바(error bar) $\Rightarrow$ 편차
- [barplot 관련 문서](http://seaborn.pydata.org/generated/seaborn.barplot.html)

```python
tips.head()
```

```python
sns.barplot(x="day", y="total_bill", data=tips)
plt.title("요일별 전체 팁")
```

### 4.4.2 `boxplot`

- `boxplot` 명령은 박스-휘스커 플롯(Box-Whisker Plot) 혹은 간단히 박스 플롯이라 부르는 차트이다.
- 박스 플롯은 박스와 박스 바깥의 선(Whisker)으로 이뤄진다.
- [boxplot 관련 문서](http://seaborn.pydata.org/generated/seaborn.boxplot.html)

- **박스**
  - 박스는 실수 값 분포에서 1사분위수(Q1)와 3사분위수(Q3)를 뜻함
  - 이 3사분위수와 1사분위수의 차이(Q3-Q1)를 **IQR(InterQuartile Range)**라고 한다.  
  
  
- **박스 내부의 가로선**
  - 중앙값을 의미  
  
  
- **박스 외부의 세로선**
  - 1사분위수보다 1.5 $\times$ IQR 만큼 낮은 값과 3사분위수보다 1.5 $\times$ IQR 만큼 높은 값의 구간을 기준으로 그 구간의 내부에 있는 가장 큰 데이터와 가장 작은 데이터를 잇는 선분  
  
  
- **그 바깥의 점**
  - 아웃라이어(outlier)
  - 일일히 점으로 표시

```python
sns.boxplot(x="day", y="total_bill", data=tips)
plt.title("요일별 전체 팁의 Box Plot")
plt.show()
```

- `barplot`과 `boxplot`은 중앙값, 표준 편차 등 분포의 간략한 특성만 보여줌
- `violinplot`, `stripplot`, `swarmplot` 등은 카테고리값에 따른 각 분포의 실제 데이터나 전체 형상을 보여준다는 장점이 있다.

### 4.4.3 `violinplot`

- `violinplot`은 세로 방향으로 커널 밀도 히스토그램을 그려준다.
- 왼쪽과 오른쪽이 대칭이 되도록하여 바이올린처럼 보인다.
- [violinplot 관련 문서](http://seaborn.pydata.org/generated/seaborn.violinplot.html)

```python
sns.violinplot(x="day", y="total_bill", data=tips)
plt.title("요일별 전체 팁의 Violin Plot")
plt.show()
```

### 4.4.4 `stripplot`

- `stripplot`은 마치 산점도처럼 모든 데이터를 점으로 그려줌
- `jitter=True`를 설정하면 가로축상의 위치를 무작위로 바꾸어서 데이터의 수가 많을 경우에 겹치지 않도록 한다.
- [stripplot 관련 문서](http://seaborn.pydata.org/generated/seaborn.stripplot.html)

```python
np.random.seed(0)
sns.stripplot(x="day", y="total_bill", data=tips, jitter=True)
plt.title("요일별 전체 팁의 Strip Plot")
plt.show()
```

### 4.4.5 `swarmplot`

- `swarmplot`은 `stripplot`과 비슷하지만 데이터를 나타내는 점이 겹쳐지지 않도록 옆으로 이동한다.
- [swarmplot 관련 문서](http://seaborn.pydata.org/generated/seaborn.swarmplot.html)

```python
sns.swarmplot(x="day", y="total_bill", data=tips)
plt.title("요일별 전체 팁의 Swarm Plot")
plt.show()
```

## 4.5 다차원 복합 데이터

- 지금까지 소개한 대부분의 명령어에는 2차원 이상의 고차원 데이터에 대해서도 분석할 수 있는 기능이 포함되어 있다.

### 4.5.1 `hue`

- `barplot`, `boxplot`, `violinplot` 등 에서는 두 가지 카테고리 값에 의한 실수값의 변화를 보기 위한 `hue` 인수를 제공한다.
- `hue` 인수에 카테고리 값을 가지는 변수의 이름을 지정하면 카테고리 값에 따라 다르게 시각화된다.
- `hue` 값이 시각화되는 방법은 플롯의 종류에 따라 다르다.  

```python
sns.barplot(x="day", y="total_bill", hue="sex", data=tips)
plt.title("요일별 성별 전체 팁의 Bar Plot")
plt.show()
```

```python
sns.boxplot(x="day", y="total_bill", hue="sex", data=tips)
plt.title("요일별 성별 전체 팁의 Box Plot")
plt.show()
```

```python
sns.violinplot(x="day", y="total_bill", hue="sex", data=tips)
plt.title("요일별 성별 전체 팁의 Violin Plot")
plt.show()
```

```python
np.random.seed(0)
sns.stripplot(x="day", y="total_bill", hue="sex", data=tips,
              jitter=True)
plt.title("요일별 성별 전체 팁의 Strip Plot")
plt.legend(loc=1)
plt.show()
```

```python
sns.swarmplot(x="day", y="total_bill", hue="sex", data=tips)
plt.title("요일별 성별 전체 팁의 Swarm Plot")
plt.legend(loc=1)
plt.show()
```

### 4.5.2 `split`, `dodge`

- `violinplot`, `stripplot`, `swarmplot` 등에서는 `split` 또는 `dodge` 옵션으로 시각화 방법을 변경할 수도 있다.

```python
sns.violinplot(x="day", y="total_bill", hue="sex", data=tips,
               split=True)
plt.title("요일별 성별 전체 팁의 Violin Plot, Split=True")
plt.show()
```

```python
sns.stripplot(x="day", y="total_bill", hue="sex", data=tips,
              jitter=True, dodge=True)
plt.title("요일별 성별 전체 팁의 Strip Plot, Split(Dodge)=True")
plt.show()
```

```python
sns.swarmplot(x="day", y="total_bill", hue="sex", data=tips,
              dodge=True)
plt.title("요일별 성별 전체 팁의 Swarm Plot, Split(Dodge)=True")
plt.show()
```

### 4.5.3 `heatmap`

- `heatmap`을 이용해도 두 개의 카테고리 값에 의한 실수 값 변화를 볼 수 있다.
- [heatmap 관련 문서](http://seaborn.pydata.org/generated/seaborn.heatmap.html)

```python
flights.head(5)
```

```python
flights_passengers = flights.pivot("month", "year", "passengers")
flights_passengers
```

```python
sns.heatmap(flights_passengers, annot=True, fmt="d",
            linewidths=1)
plt.title("연도, 월별 승객수에 대한 Heatmap")
plt.show()
```

# 5. `catplot`

- `catplot`은 색상(hue)과 행(row) 등을 동시에 사용하여 3개 이상의 카테고리 값에 의한 분포 변화를 보여준다.
- [catplot 관련 문서](http://seaborn.pydata.org/generated/seaborn.factorplot.html)

```python
data = titanic[titanic.survived.notnull()]
```

```python
data.head(3)
```

```python
sns.catplot(x="age", y="sex", hue="survived", row="class",
            data=data, kind="violin", split=True,
            height=2, aspect=4)
```

```python
sns.catplot(x="sex", y="age", hue="survived", row="class",
            data=data, kind="violin", split=True,
            height=3, aspect=2)
```

```python
sns.catplot(x="sex", y="age", hue="survived", row="class",
            data=data, kind="strip", split=True,
            height=3, aspect=2)
```

```python
sns.catplot(x="sex", y="age", hue="survived", row="class",
            data=data, kind="swarm", dodge=True,
            height=3, aspect=2)
```
