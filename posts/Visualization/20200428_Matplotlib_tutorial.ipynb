# 1. 처음으로 그려보는 그래프

```python
import matplotlib
%matplotlib inline
# matplotlib.use("TKAgg") # 그래픽 백엔드로 Tk를 사용하고자 한다면, 이 코드를 사용
```

## 1.1 `plt.plot()`

```python
import matplotlib.pyplot as plt

plt.plot([1, 2, 4, 9, 5, 3])
plt.show()
```

```python
plt.plot([-3, -2, 5, 0], [1, 6, 4, 3])
plt.show()
```

<br>

## 1.2 `plt.axis()`

```python
# axis 함수 이용 축의 범위 조절
plt.plot([-3, -2, 5, 0], [1, 6, 4, 3])
plt.axis([-4, 6, 0, 7])
plt.show()
```

```python
import numpy as np

x = np.linspace(-2, 2, 500)
print(x[:5])
```

```python
y = x**2
print(y[:5])
```

```python
plt.plot(x, y)
plt.show()
```

<br>

## 1.3 `plt.title()`, `plt.xlabel()`, `plt.ylabel()`, `plt.grid()`

```python
plt.plot(x, y)
plt.title("Square function")
plt.xlabel("x")
plt.ylabel("y = x**2")
plt.grid(True)
plt.show()
```

# 2. 선의 스타일과 색상

기본적으로 matplotlib은 바로 다음에 위치한(연이은) 데이터 사이에 선을 그립니다.

```python
plt.plot([0, 100, 100, 0, 0, 100, 50, 0, 100], [0, 0, 100, 100, 0, 100, 130, 100, 0])
plt.axis([-10, 110, -10, 140])
plt.show()
```

<br>

## 2.1 `plt.plot(x, y, 선스타일 및 색상 지정)`

```python
plt.plot([0, 100, 100, 0, 0, 100, 50, 0, 100], [0, 0, 100, 100, 0, 100, 130, 100, 0], "g--")
plt.axis([-10, 110, -10, 140])
plt.show()
```

```python
plt.plot([0, 100, 100, 0, 0], [0, 0, 100, 100, 0], "r-", 
         [0, 100, 50, 0, 100], [0, 100, 130, 100, 0], "g--")
plt.axis([-10, 110, -10, 140])
plt.show()
```

```python
plt.plot([0, 100, 100, 0, 0], [0, 0, 100, 100, 0], "r-")
plt.plot([0, 100, 50, 0, 100], [0, 100, 130, 100, 0], "g--")
plt.axis([-10, 110, -10, 140])
plt.show()
```

<br>

## 2.2 선스타일 및 색상 지정 옵션

### 2.2.1 선 스타일 옵션

| character | description           |
| --------- | --------------------- |
| `'-'`     | solid line style      |
| `'--'`    | dashed line style     |
| `'-.'`    | dash-dot line style   |
| `':'`     | dotted line style     |
| `'.'`     | point marker          |
| `','`     | pixel marker          |
| `'o'`     | circle marker         |
| `'v'`     | triangle_down marker  |
| `'^'`     | triangle_up marker    |
| `'<'`     | triangle_left marker  |
| `'>'`     | triangle_right marker |
| `'1'`     | tri_down marker       |
| `'2'`     | tri_up marker         |
| `'3'`     | tri_left marker       |
| `'4'`     | tri_right marker      |
| `'s'`     | square marker         |
| `'p'`     | pentagon marker       |
| `'*'`     | star marker           |
| `'h'`     | hexagon1 marker       |
| `'H'`     | hexagon2 marker       |
| `'+'`     | plus marker           |
| `'x'`     | x marker              |
| `'D'`     | diamond marker        |
| `'d'`     | thin_diamond marker   |
| `'|'`     | vline marker          |
| `'_'`     | hline marker          |

<br>

### 2.2.2 선 색상 옵션

| character | color   |
| --------- | ------- |
| `b`       | blue    |
| `g`       | green   |
| `r`       | red     |
| `c`       | cyan    |
| `m`       | magenta |
| `y`       | yellow  |
| `k`       | black   |
| `w`       | white   |

```python
x = np.linspace(-1.4, 1.4, 30)
plt.plot(x, x, 'g--', # 초록색 파선
         x, x**2, 'r:', # 빨간색 점선
         x, x**3, 'b^') # 파란색 삼각형
plt.show()
```
